```python
import os
path = '/data/nmt'
```

```python
output_filename = os.path.join(path, 'corpus.txt')
corpus = [('jhe-koen-dev.en', 'jhe-koen-dev.ko'),
('jhe-koen-eval.en', 'jhe-koen-eval.ko'),
('korean-english-park.dev.en', 'korean-english-park.dev.ko'),
('korean-english-park.test.en', 'korean-english-park.test.ko'),
('korean-english-park.train.en', 'korean-english-park.train.ko')]
```

```python
all_en_sents = []
all_kr_sents = []
for cor in corpus:
    (en, kr) = cor
    
    en_sents = []
    with open(os.path.join(path, en), 'r') as f:
        for oneline in f:
            sent = oneline.rstrip()
            en_sents.append(sent)
            
    kr_sents = []
    with open(os.path.join(path, kr), 'r') as f:
        for oneline in f:
            sent = oneline.rstrip()
            kr_sents.append(sent)
    
    all_en_sents += en_sents
    all_kr_sents += kr_sents
    print('{} -> {} and {}'.format(cor, len(en_sents), len(kr_sents)))
    
```

```python
with open(output_filename, 'w') as f:
    for en, kr in zip(all_en_sents, all_kr_sents):
        f.write('\t'.join([en, kr]) + '\n')
```
